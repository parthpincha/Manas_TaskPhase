```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Sigmoid function
# -----------------------------
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


# -----------------------------
# Log Loss function
# -----------------------------
def log_loss(y_true, y_pred):
    epsilon = 1e-15  # to avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(
        y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)
    )


# -----------------------------
# Sample data
# x_values -> input feature
# y_values -> true labels (0 or 1)
# -----------------------------
x_values = np.array([18, 20, 22, 25, 30])
y_values = np.array([0, 0, 0, 1, 1])

# -----------------------------
# Initialize parameters
# -----------------------------
w = 0.1
b = 0.0
learning_rate = 0.01
iterations = 100

# -----------------------------
# Gradient Descent
# -----------------------------
loss_history = []

for i in range(iterations):
    z = w * x_values + b          # linear model
    y_pred = sigmoid(z)           # prediction

    loss = log_loss(y_values, y_pred)
    loss_history.append(loss)

    # gradients
    dw = np.mean((y_pred - y_values) * x_values)
    db = np.mean(y_pred - y_values)

    # update parameters
    w -= learning_rate * dw
    b -= learning_rate * db

    if i % 10 == 0:
        print(f"Iteration {i}: Log Loss = {loss}")

# -----------------------------
# Plot Log Loss vs Iterations
# -----------------------------
plt.figure()
plt.plot(loss_history)
plt.xlabel("Iterations")
plt.ylabel("Log Loss")
plt.title("Log Loss Minimization using Gradient Descent")
plt.show()

# -----------------------------
# Plot Sigmoid Function
# -----------------------------
z_range = np.linspace(-10, 10, 100)
plt.figure()
plt.plot(z_range, sigmoid(z_range))
plt.xlabel("z")
plt.ylabel("sigmoid(z)")
plt.title("Sigmoid (Logistic) Function")
plt.show()
```
