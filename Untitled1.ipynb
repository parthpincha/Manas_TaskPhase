```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms

transform = transforms.ToTensor()

train_dataset = torchvision.datasets.MNIST(root="./data", train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.MNIST(root="./data", train=False, download=True, transform=transform)

train_dataloader = torch.utils.data.DataLoader(train_dataset,batch_size = 64, shuffle=True)
test_dataloader = torch.utils.data.DataLoader(test_dataset,batch_size = 64, shuffle=False)

print("dataset downloaded batched and loaded and ready")
```

```python
class MNIST_MLP(nn.Module):
    def __init__(self,input_size,h1,h2,output_size):
        super(MNIST_MLP,self).__init__()
    
        self.layer_1 = nn.Linear(input_size,h1)
        self.layer_2 = nn.Linear(h1,h2)
        self.layer_3 = nn.Linear(h2,output_size)

        self.relu = nn.ReLU()

    def forward(self,x):
        
        x=x.view(-1,28*28)

        x = self.relu(self.layer_1(x))
        x = self.relu(self.layer_2(x))
        x = self.layer_3(x)

        return x

model = MNIST_MLP(784, 128, 32, 10)

criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(model.parameters(),lr=0.001)

epochs = 12

for i in range(epochs):
    running_loss = 0.0

    for images, labels in train_dataloader: # Use plural to stay consistent
        
        optimizer.zero_grad()

        outputs = model(images) # 'images' matches the loop variable

        loss = criterion(outputs, labels) # 'labels' matches the loop variable

        loss.backward()

        optimizer.step()

        running_loss += loss.item() # Don't forget to actually add the loss to the total!

    # Move this print out of the inner loop
    print(f"Epoch {i+1} | Loss: {running_loss/len(train_dataloader):.4f}")
```

```python
# 5. EVALUATION
model.eval() # Turn off Dropout/BatchNorm for testing
correct = 0
total = 0

with torch.no_grad(): # Disable gradient tracking to save memory
    for images, labels in test_dataloader: # Note: you'd need the test_loader from earlier
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Final Accuracy on 10,000 test images: {100 * correct / total:.2f}%")
```
