```python
import numpy as np
import random
from math import sqrt

# decleare 2 random arrays with x and y values
def arrayDeclearation(n):
    x_values = np.random.randint(0,50,n)
    x1_values = np.random.randint(0,50,n)
    y_values = ((x_values + x1_values)//2 + np.random.randint(-10,10,n))
    return x_values, x1_values, y_values


x_values, x1_values, y_values = arrayDeclearation(10)

print(x_values)
print(x1_values)
print(y_values)

```

```python
# predicting values for slope and intercept

def findingSlopeIntercept(x_values,x1_values, y_values):
    x_mean = np.mean(x_values)
    x1_mean = np.mean(x1_values)
    y_mean = np.mean(y_values)

    numerator_x = np.dot((x_values - x_mean),(y_values - y_mean))
    denominator_x = np.sum((x_values - x_mean)**2)

    slope_x = numerator_x / denominator_x

    numerator_x1 = np.dot((x1_values - x1_mean),(y_values - y_mean))
    denominator_x1 = np.sum((x1_values - x1_mean)**2)

    slope_x1 = numerator_x1 / denominator_x1
    
    intercept = y_mean - slope_x*x_mean - slope_x1*x1_mean

    return slope_x, slope_x1, intercept

slope_x, slope_x1, intercept = findingSlopeIntercept(x_values, x1_values, y_values)

y_pred_values = np.array(intercept + slope_x*x_values + slope_x1*x1_values)
print(y_pred_values)  
```

```python
#calcualting the root mean sqaure error

def rootMeanSquareError(actual_values, predicted_values):
    sum_error = 0.0
    for y, y_pred in zip(actual_values, predicted_values):
        pred_error = y_pred - y
        sum_error += (pred_error**2)
    mean_error = sum_error/float(len(actual_values))
    return sqrt(mean_error)

# _________________________________--replace x values with predicted____________________-
rmse = rootMeanSquareError(y_values, y_pred_values)

print(rmse)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_values, y_values, label='Observed Value')
plt.plot(x_values, y_pred_values, label='Predicted Value', color='red')
plt.xlabel('<--X-Axis-->')
plt.ylabel('<--Y-Axis-->')
plt.legend()
plt.show()
```

The goal of this implementation was to extend simple linear regression to multiple input variables by modeling the output as a linear combination of multiple predictors and a shared intercept. Conceptually, this corresponds to fitting a plane in three-dimensional space that best represents the relationship between the inputs and the target variable.

To achieve this, separate slopes were computed for each predictor using the covarianceâ€“variance formula from simple linear regression. Each predictor was regressed independently against the target variable, and the resulting slopes were combined to generate predictions. A mean-based intercept was then calculated to anchor the combined model to the center of the data.

This approach was intended to account for the contribution of each variable individually and to combine their effects into a single prediction. However, this method fails to perform true multivariable linear regression because each slope is estimated in isolation, without accounting for correlations between predictors. As a result, shared variance among input variables is not properly resolved, leading to distorted slope estimates that do not minimize the joint mean squared error.

Additional attempts to correct this behavior by adjusting the intercept or modifying the prediction formula were ineffective, as the overlap between predictors arises during slope estimation rather than at the prediction or intercept stage. Geometrically, the implementation fits multiple independent regression lines in orthogonal projections and combines them, rather than fitting a single plane directly to the data.

Although the model produces reasonable-looking predictions for some datasets, it does not represent the statistically correct solution to multivariable linear regression.
