```python
#___________________________________from csv file________________________________________________
import numpy as np
import random
import pandas as pd
from math import sqrt

# decleare 2 random arrays with x and y values
df = pd.read_csv(r"C:\Users\Parth_Coding\Desktop\weatherHistory.csv")

x_column_name = "Temperature (C)"
y_column_name = "Humidity"

x_values = df[x_column_name].to_numpy()[:100]
y_values = df[y_column_name].to_numpy()[:100]


print(x_values)
print(y_values)
```

```python
# intialise weights randomly

def initialiseWeights():
    w0 = np.random.rand()
    w1 = np.random.rand()
    return w0,w1

# cost calculation
def costCaluculation(x_values,y_values,y_pred):
    cost = 1/(2*(len(y_values))) * np.sum(np.square(y_pred - y_values))
    return cost
    

# calculate the y_predicted values
def calculateYPredicted(x_values, w1,w0):
    return np.dot(x_values,w1)+w0


#updating the w0 and w1 parameters
def update_parameters(x_values,y_values,y_pred,cost,w0,w1,learning_rate):
  m = len(y_values)
  db=(np.sum(y_pred-y_values))/m
  dw=np.dot(y_pred - y_values,x_values)/m

  w0_new=w0-learning_rate*db
  w1_new=w1-learning_rate*dw
  return w0_new,w1_new
```

```python
# main loop of gradient descent for set number of itreations
import matplotlib.pyplot as plt

cost_history = []
iterations = 15
learning_rate = .001


w0,w1 = initialiseWeights()


for i in range (0,iterations):
    y_pred = calculateYPredicted(x_values, w1,w0)
    cost = costCaluculation(x_values,y_values,y_pred)
    w0,w1 = update_parameters(x_values,y_values,y_pred,cost,w0,w1,learning_rate)
    cost_history.append(cost)
```

```python
plt.figure()
plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost vs Iterations (Gradient Descent)")
plt.show()
```
